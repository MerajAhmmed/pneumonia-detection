```python
import os
from keras.preprocessing.image import ImageDataGenerator
from keras.preprocessing import image
import matplotlib.pyplot as plt
from keras.utils import load_img, img_to_array
```

```python
# Set the path to the folder containing your original images
original_folder_path = 'E:\\Thesis Related Work\\Research\\Pneumonia_Detection\\Pneumonia'

# Set the path to the folder where you want to save augmented images
augmented_folder_path = 'E:\\Thesis Related Work\\Research\\Pneumonia_Detection\\Augmented pneumonia'
```

```python
# Create an ImageDataGenerator instance with the specified augmentations
datagen = ImageDataGenerator(
    rotation_range=15,
    zoom_range=0.1,
    horizontal_flip=False,
    vertical_flip=False,
    brightness_range=[0.2, 0.3],
    fill_mode='constant',  # Set fill_mode to 'constant'
    cval=0  # Set cval to 0 for black color
)
```

```python
# Get the list of image files in the original folder
image_files = [f for f in os.listdir(original_folder_path) if os.path.isfile(os.path.join(original_folder_path, f))]
```

```python
# Create the augmented images and save them to the augmented folder
for image_file in image_files:
    img_path = os.path.join(original_folder_path, image_file)
    img = load_img(img_path)
    x = img_to_array(img)
    x = x.reshape((1,) + x.shape)

    # Generate augmented images
    i = 0
    for batch in datagen.flow(x, batch_size=16, save_to_dir=augmented_folder_path, save_prefix='aug', save_format='jpeg'):
        i += 1
        if i >= 10:  # Generate 5 augmented images per original image
            break
```
